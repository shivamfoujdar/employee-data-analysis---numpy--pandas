```python
import numpy as np
import pandas as pd

np.random.seed(1000)
rows =5000000
data = {
    "Id" : np.arange(1,rows+1),
    "Age": np.random.randint(22,60,size=rows),
    "Salary": np.random.randint(30000,150000,size=rows),
    "Experienced_years":np.round(np.random.uniform(1,15,size=rows),1),
    "Department":np.random.choice(["IT","HR","FINANCE","MARKETING","SALES"],size=rows),
    "Perfomance_Score":np.round(np.random.uniform(1,5,size=rows),1),
    "target":np.random.choice([0 ,1],size=rows)
    
    
}

df=pd.DataFrame(data)
df.to_csv(r"E:\output.csv", index=False)



df.head()



```
